```python
!pip install ipywidgets --quiet

import json
import random
import ipywidgets as widgets
from IPython.display import display, HTML

# STEP 1: Save intents.json
intents_data = {
    "intents": [
        {
            "tag": "greeting",
            "patterns": ["Hi", "Hello", "Hey"],
            "responses": ["Hello!", "Hi there!", "Hey! How can I help you?"]
        },
        {
            "tag": "goodbye",
            "patterns": ["Bye", "See you later"],
            "responses": ["Goodbye!", "See you soon!"]
        },
        {
            "tag": "thanks",
            "patterns": ["Thanks", "Thank you"],
            "responses": ["You're welcome!", "No problem!"]
        },
        {
            "tag": "name",
            "patterns": ["What is your name?", "Who are you?"],
            "responses": ["I'm your smart AI chatbot 😎", "You can call me ChatBot-X!"]
        }
    ]
}

with open("intents.json", "w") as f:
    json.dump(intents_data, f, indent=4)

# STEP 2: Load intents
with open("intents.json") as file:
    data = json.load(file)

# STEP 3: Chatbot logic
def chatbot_response(msg):
    for intent in data['intents']:
        if msg.lower() in [p.lower() for p in intent['patterns']]:
            return random.choice(intent['responses'])
    return "🤖 Sorry, I didn’t get that."

# STEP 4: Create UI
chat_box = widgets.HTML(
    value="<b>🤖 ChatBot-X is ready to chat!</b><br>",
    layout=widgets.Layout(border='1px solid #ccc', padding='10px', width='100%', height='300px', overflow_y='scroll', background_color='#fff')
)

user_input = widgets.Text(
    placeholder='Type your message here...',
    layout=widgets.Layout(width='85%')
)

send_button = widgets.Button(
    description='Send',
    button_style='primary',
    layout=widgets.Layout(width='13%', margin='0 0 0 5px')
)

input_box = widgets.HBox([user_input, send_button])

chat_ui = widgets.VBox([chat_box, input_box])
display(chat_ui)

# STEP 5: Send message handler
def send_message(_):
    msg = user_input.value.strip()
    if msg == "":
        return
    bot_reply = chatbot_response(msg)
    chat_box.value += f"<b>You:</b> {msg}<br><b>Bot:</b> {bot_reply}<br><br>"
    user_input.value = ""

send_button.on_click(send_message)
user_input.on_submit(send_message)
```
